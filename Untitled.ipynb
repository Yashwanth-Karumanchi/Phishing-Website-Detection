```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('pages/data/image_embeds.csv')

# Display the DataFrame before replacement
print("Before replacement:")
print(df)

# Define the value to replace
old_value = 'exampleabc'

# Define the new value
new_value = 'google-search'

# Replace the value in the specified cell
df.loc[df['Domain'] == old_value, 'Domain'] = new_value

# Display the DataFrame after replacement
print("After replacement:")
print(df)

# Save the DataFrame back to the CSV file
df.to_csv('pages/data/image_embeds.csv', index=False)
```
